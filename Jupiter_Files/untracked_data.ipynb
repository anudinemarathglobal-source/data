```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from gspread_pandas import Spread, conf
```

```python
config_path = r'C:\Users\USER\Downloads\Emarath_global\service_account.json'
# 3. Define the Sheet URL (Name - EMARATH GLOBAL -DASHBOARD)
emarath_global_sheet_url = "https://docs.google.com/spreadsheets/d/1aXobnCl4QQ-hQ_IWoTPo2ZEo3HM5kNwycwMBJtEDOfc/edit?gid=822921427#gid=822921427"
# Load the config from your service account JSON
c = conf.get_config(file_name=config_path)

# Connect using that config object
emarath_spread = Spread(emarath_global_sheet_url, config=c)
```

```python
master_df = pd.read_excel(r'C:\Users\USER\Downloads\Emarath_global\Data\flowbee_detaild_report.xlsx', skiprows=4)
master_df
```

```python
# 5. Loop through and pull categorized data
all_data_list = []
all_sheets = emarath_spread.sheets

# Define exactly which statuses you want to track
target_date = "2026-11-02" 

for sheet in all_sheets:
    sheet_name = sheet.title
    
    if sheet_name in ["MAIN DASHBOARD", "CRM", "Dropdowns", "Sheet1", "DASHBOARD", "20"]:
        continue
    
    try:
        df = emarath_spread.sheet_to_df(sheet=sheet_name, index=None, header_rows=1)
        
        if not df.empty and all(col in df.columns for col in ['DATE', 'NAME', 'PHONE NO 1']):
      
            df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')

            mask = df['DATE'].dt.date == pd.to_datetime(target_date).date()
            filtered_df = df[mask].copy()
            
            if not filtered_df.empty:
                filtered_df['Name of BDE'] = sheet_name
                filtered_df['DATE'] = filtered_df['DATE'].dt.strftime('%Y-%d-%m')
                all_data_list.append(filtered_df[['DATE', 'Name of BDE', 'NAME', 'PHONE NO 1']])
            
    except Exception as e:
        print(f"Skipping {sheet_name}: {e}")


    
```

```python
final_df = final_df.dropna(subset=['NAME', 'PHONE NO 1'])


final_df['PHONE NO 1'] = final_df['PHONE NO 1'].astype(str).str.strip()
master_df['MOBILE'] = master_df['MOBILE'].astype(str).str.strip()

# 3. Perform the Anti-Join
# We use 'left' to keep final_df records and 'indicator' to track matches
merged = final_df.merge(
    master_df[['MOBILE']], 
    left_on=['PHONE NO 1'], 
    right_on=[ 'MOBILE'], 
    how='left', 
    indicator=True
)


new_leads_df = merged[merged['_merge'] == 'left_only'].copy()

new_leads_df = new_leads_df.drop(columns=['_merge',  'MOBILE'])

new_leads_df = new_leads_df.drop_duplicates(subset=[ 'PHONE NO 1'])

print(f"Isolated {len(new_leads_df)} unique records not found in Master Data.")

# 7. Export to Excel
new_leads_df.to_excel("isolated_new_leads.xlsx", index=False)
```
